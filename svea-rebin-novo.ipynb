```python
import matplotlib.pyplot
import matplotlib.pyplot as plt
import cana
from cana.datasets import getspectrum
import numpy as np
import plotly.express as px

df = px.data.tips()

spec = cana.loadspec('/home/camila/Desktop/TCC/TCC-Asteroides-jupyter-notebook/svea.txt')

spec = spec.mask_region(region=[2.55,2.85])
spec = spec.clean_spec()

spec = spec.trim(0.7,4.0) 

spec_ir1 = spec
spec_ir2 = spec.trim(2.4,4.2)

aux = np.where(spec.w > 1.00)[0] 
spec_ir2 = spec[aux] 

spec2_21bin_rebin = spec_ir2.rebin(21)

specnovo21 = cana.stack_spec([spec, spec2_21bin_rebin])

specnovo21.plot()

















spec = spec.trim(0.7,4.0) 

spec = spec.clean_spec()

aux = np.where(spec.w < 2.4)[0] 
spec_ir1 = spec[aux] 


aux = np.where(spec.w > 2.4)[0] 
spec_ir2 = spec[aux] 

spec_ir2 = spec.trim(2.4,4.2)

specnew = cana.stack_spec([spec_ir1, spec_ir2])

aux = np.where(specnew.w < 2.4)[0] 
spec1 = specnew[aux] 

aux = np.where(specnew.w > 2.4)[0] 
spec2 = specnew[aux] 

spec2_21bin_rebin = spec2.rebin(21)

specnovo21 = cana.stack_spec([spec1, spec2_21bin_rebin])

specnovo21.plot()








matplotlib.pyplot.title("Asteroide 329 Svea com binagem")

matplotlib.pyplot.xlabel("Comprimento de onda")
matplotlib.pyplot.ylabel("Reflectância")

plt.plot(specnovo21.w, specnovo21.r,"black",lw=1.0,linestyle="solid")


plt.figure(figsize=(20, 12))

specnovo21.save("329-Svea-com-binagem.txt")
```
